```r
library(tidyverse)
H<-NULL
for(i in 1:5){
    C<-readline(prompt="Enter box color (0 for White / 1 for Black): ") %>% as.numeric()
    H<-c(H, C)
}
iter<-readline(prompt="Enter the number of extractions: ") %>% as.numeric()




White<-function(AH){
    BH<-NULL
    i<-0:5
    AE<-sum((i/5)*AH[i+1])
    for(j in 0:5){
        val<-((j/5)*AH[j+1])/AE
        BH<-c(BH, val)
    }
    return(BH)
}
Black<-function(AH){
    BH<-NULL
    i<-0:5
    AE<-sum(((5-i)/5)*AH[i+1])
    for(j in 0:5){
        val<-(((5-j)/5)*AH[j+1])/AE
        BH<-c(BH, val)
    }
    return(BH)
}



PH<-rep(1/6, 6)

set.seed(89540)
Random<-sample(1:5, size=iter, replace=TRUE)     
cat("\nH0                  H1                 H2                H3                 H4                H5")
PH
for(i in 1:iter){
    if(H[Random[i]]==1){
        PH<-Black(PH) 
    }else{
        PH<-White(PH)
    }
    cat("\n" , PH)
}



```
